```python
#from a pandas dataframe
batsman=pd.read_csv('sharma-kohli.csv')
batsman

plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
#Plotting multiple plots
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.show()
```

```python
# labels title
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])

plt.title('Sharma vs kohli career comparison')
plt.xlabel('Season')
plt.ylabel('Runs')


```

```python
# colors(hex) and line(width & style) and marker(size)
plt.plot(batsman['index'],batsman['V Kohli'],color='green')
plt.plot(batsman['index'],batsman['RG Sharma'],color='black')

plt.title('Sharma vs kohli career comparison')
plt.xlabel('Season')
plt.ylabel('Runs scored')
```

```python
#linestyle can be- dashed,solid,dotted,dashdot
#marker='d' --(diamond),+,.

plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='dashdot',linewidth=2,marker='d')
plt.plot(batsman['index'],batsman['RG Sharma'],color='pink',linestyle='solid',linewidth=3,marker='o',markersize=7)

plt.title('Sharma vs kohli career comparison')
plt.xlabel('Season')
plt.ylabel('Runs scored')
```

```python
# legend-location
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='dashdot',linewidth=2,marker='d',label='virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='pink',linestyle='solid',linewidth=3,marker='o',markersize=7,label='sharma')

plt.title('Sharma vs kohli career comparison')
plt.xlabel('Season')
plt.ylabel('Runs scored')

plt.legend(loc='upper center')
```

```python
# limiting axes
price=[48000,54000,57000,49000,47000,45000,4500000]
year=[2015,2016,2017,2018,2019,2020,2021]
plt.plot(year,price)

plt.ylim(0,75000)
plt.xlim(2017,2019)
```

```python
# Grid
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='dashdot',linewidth=2,marker='d',label='virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='pink',linestyle='solid',linewidth=3,marker='o',markersize=7,label='sharma')

plt.title('Sharma vs kohli career comparison')
plt.xlabel('Season')
plt.ylabel('Runs scored')
plt.grid()
plt.legend(loc='upper center')

plt.show()
```

3D SURFACE PLOTS

```python
x=np.linspace(-10,10,100)
y=np.linspace(-10,10,100)
xx,yy=np.meshgrid(x,y)
z=xx**2+yy**2

fig=plt.figure(figsize=(12,8))
ax=plt.subplot(projection='3d')
p=ax.plot_surface(xx,yy,z,cmap='winter')
fig.colorbar(p)
```

```python
z=np.sin(xx) + np.cos(yy)

fig=plt.figure(figsize=(12,8))
ax=plt.subplot(projection='3d')
p=ax.plot_surface(xx,yy,z,cmap='winter')
fig.colorbar(p)
```

Contour Plots

```python
x=np.linspace(-10,10,100)
y=np.linspace(-10,10,100)
xx,yy=np.meshgrid(x,y)
z=xx**2+yy**2

fig=plt.figure(figsize=(12,8))
ax=plt.subplot()
p=ax.contour(xx,yy,z,cmap='winter')
fig.colorbar(p)
```

```python
fig=plt.figure(figsize=(12,8))
ax=plt.subplot()
p=ax.contourf(xx,yy,z,cmap='winter')
fig.colorbar(p)
```
