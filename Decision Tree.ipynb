## **What is entropy??**
Entropy is nothing but measure of disorder.
<br>
*More knowledge less entropy*
<br>

Mathematical Formula - 
<br>
$$E(S)=\sum_{i=1}^{c}-p_i \log_2 p_i$$

## **OBSERVATION**
1) More the uncertainity more is entropy
2) For a 2 class problem the min entropy is 0 and the max is 1
3) For more than 2 classes the min entropy is 0 but the max can be >1
4) Both log_2 or log_e can be used to calculate entrop

## **INFORMATION GAIN**
It is based on the decrease in entropy after a dataset is split on an attribute.
<BR>
Contructing a decision tree is all abt finding attribute that returns the highest information gain.
<br>

Information gain = E(Parent) - {Weighted Average}*E{Children}
<br>
Calculate information gain for all the columns
<br>
*Whichever column has the highest information gain(max dec in entropy) the algorithm will select that column to split the data*
<br>
Decision tree then applies a recursive greedy search algorithm in top bottom fashion to find information Gain at every level of the tree.
<br>
Once a leaf node is reached (Entropy=0),no more splitting is done.

## **Overfitting & Underfitting**

### **OVERFITTING**
Decision tree performs great on training data but performs poorly on test data — that's overfitting. The model has basically memorized the training data instead of learning general patterns.

- Performs well → Training
- Performs badly → Test

If you set `max_depth = None`, the tree keeps splitting until every leaf is pure (or has very few samples) — this makes the tree very deep and complex, giving it a high chance of overfitting.

### **UNDERFITTING**
If you set `max_depth = 1`, the tree makes only one split — meaning it barely looks at the data before making a prediction. It hasn't learned enough patterns to be useful.

- This is underfitting — the model is too simple, performs poorly on BOTH training and test data (not just test).

**So we don't want either extreme.**

**Kya issue hai:** Max_depth zyada bhi nahi rakhna jisse overfitting ho jaye, aur itna kam bhi nahi rakhna jisse underfitting ho jaye — beech ka sahi balance dhoondna hai (this is called the bias-variance tradeoff).

### **HYPERPARAMETERS**

- **`max_depth`** — controls how deep the tree can grow. Lower = simpler tree (risk underfitting), higher = complex tree (risk overfitting).

- **`splitter`** — set to `'random'` instead of default `'best'`. Instead of always picking the single best split, it picks a random one among good options — adds randomness, which reduces overfitting (similar spirit to what random forest does).

- **`min_samples_split`** — minimum number of samples a node must have before it's allowed to split further. E.g., if set to 100, a node with fewer than 100 samples won't split anymore — this stops the tree from making super-specific splits on tiny groups (which usually means memorizing noise).

- **`min_samples_leaf`** — minimum number of samples that must exist in a leaf (the final end node) after a split. This prevents the tree from creating leaves with just 1-2 samples, which are usually overfit to specific rows rather than general patterns.

- **`max_features`** — limits how many columns the tree is allowed to consider when looking for the best split at each step (instead of considering all columns every time). Adds randomness, reduces overfitting — this is actually the exact trick random forest uses across its many trees.

- **`max_leaf_nodes`** — caps the total number of leaf (end) nodes the tree can have overall. Fewer leaf nodes = simpler tree = less overfitting risk.

- **`min_impurity_decrease`** — a split only happens if it reduces impurity (Gini/entropy) by at least this amount. If you increase this value, fewer splits happen (a split has to "prove" it's worth doing), which reduces overfitting.
